```python
import wordcloud
import jieba
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# w
w = wordcloud.WordCloud(width=1000,
                        height=700,
                        background_color='white',
                        font_path='msyh.ttc')

# string
f = open('words\\.txt',encoding='utf-8')
txt = f.read()
txtlist = jieba.lcut(txt)
string = " ".join(txtlist)

# stringwgenerate()
w.generate(string)

# 
w.to_file('WordClouds/output5-village.png')
img=Image.open("wordclouds/output5-village.png")
m=np.asarray(img)
plt.imshow(m)
```
