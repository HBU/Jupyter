```python
import wordcloud
import jieba
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import imageio
```

```python
mk = imageio.imread("picture/chinamap.png")
w = wordcloud.WordCloud(mask=mk)

# wscale
w = wordcloud.WordCloud(width=1000,
                        height=700,
                        background_color='white',
                        font_path='msyh.ttc',
                        mask=mk,
                        scale=15)

# string
f = open('words/.txt',encoding='utf-8')
txt = f.read()
txtlist = jieba.lcut(txt)
string = " ".join(txtlist)

# stringwgenerate()
w.generate(string)

# 
w.to_file('WordClouds/output7-chinamap.png')
img=Image.open("wordclouds/output7-chinamap.png")
m=np.asarray(img)
plt.imshow(m)
```
